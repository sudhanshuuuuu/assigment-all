```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
df = pd.read_csv('/content/549/tips.csv')
# Generate the graphs
# 1. Histogram of total_bill
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='total_bill', kde=True)
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Count')
plt.show()
# 2. Bar plot of average tip by day
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='day', y='tip')
plt.title('Average Tip by Day')
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.show()
# 3. Box plot of total_bill by time and sex
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='time', y='total_bill', hue='sex')
plt.title('Total Bill by Time and Sex')
plt.xlabel('Time')
plt.ylabel('Total Bill')
plt.legend(title='Sex')
plt.show()
# 4. Scatter plot of total_bill vs. tip
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='total_bill', y='tip')
plt.title('Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
# 5. Violin plot of total_bill by day and time
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='day', y='total_bill', hue='time',
split=True)
plt.title('Total Bill by Day and Time')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.legend(title='Time')
plt.show()
# 6. Line plot of average tip by size
plt.figure(figsize=(8, 6))
sns.lineplot(data=df, x='size', y='tip')
plt.title('Average Tip by Party Size')
plt.xlabel('Party Size')
plt.ylabel('Average Tip')
plt.show()
# 7. Bar plot of total_bill by smoker
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='smoker', y='total_bill')
plt.title('Total Bill by Smoker')
plt.xlabel('Smoker')
plt.ylabel('Total Bill')
plt.show()
# 8. Pie chart of the percentage of meals by time
plt.figure(figsize=(8, 6))
df['time'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage of Meals by Time')
plt.ylabel('')
plt.show()
# 9. Heatmap of correlations between numerical variables
plt.figure(figsize=(8, 6))
sns.heatmap(df[['total_bill', 'tip', 'size']].corr(), annot=True,
cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
# 10. Count plot of the number of meals by day and time
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='day', hue='time')
plt.title('Number of Meals by Day and Time')
plt.xlabel('Day')
plt.ylabel('Count')
plt.legend(title='Time')
plt.show()
```
